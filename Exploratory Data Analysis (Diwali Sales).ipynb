# Exploratory Data Analysis

Gender

```python
df.columns
```

```python
dx = sns.countplot(x='Gender', data = df)
for bars in dx.containers:
    dx.bar_label(bars)
```

From the above graph the count of females is more than men.

```python
# Bar Chart of Gender vs Amount
sales_gen = df.groupby(["Gender"], as_index=False )["Amount"].sum().sort_values( by='Amount',ascending=False)
sns.barplot(x="Gender", y="Amount", data=sales_gen)
```

From the above bar graph we can see that most of the buyers are females than mens and females spends more as compared to mens.

# Age


```python
dx = sns.countplot(x="Age Group", data = df, hue = "Gender")
for bars in dx.containers:
    dx.bar_label(bars)
```

```python
# Bar chart of Age Group vs Amount
sales_age = df.groupby(["Age Group"], as_index = False)["Amount"].sum().sort_values( by="Amount", ascending=False)
sns.barplot(x= "Age Group", y= "Amount", data= sales_age)
```

From the above graph we can see that most of the buyers are of the Age group between 26-35.

# State

```python
# No. of orders from 10 states
sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.barplot(x="State", y="Orders", data=sales_state)
sns.set({"figure.figsize":(25,5)})
```

```python
# amount of sales from top 10 states
sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
sns.barplot(x="State", y="Amount", data=sales_state)
sns.set({"figure.figsize":(25,5)})
```

from the above graphs we can say that most of the orders and sales amount are from UP, Maharastra and karnataka.

# Marital Status

```python
dx=sns.countplot(x="Marital_Status",data=df)
for bars in dx.containers:
    dx.bar_label(bars)
sns.set({'figure.figsize':(10,5)})
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set({'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

from the above graph we can say that most of the buyers are married womens and they spend more than men

# Occupation

```python
dx = sns.countplot(data = df, x = 'Occupation')
for bars in dx.containers:
    dx.bar_label(bars)
sns.set({'figure.figsize':(30,5)})
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set({'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount')
```

from the above bar graph we can say that most of the buyers are working in IT, Healthcare and aviation sector.

# Product Category 

```python
sns.set({'figure.figsize':(20,5)})
dx = sns.countplot(data = df, x = 'Product_Category')

for bars in dx.containers:
    dx.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set({'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount')
```

```python
# Top 10 most sold products
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)
sns.set({'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders')
```
